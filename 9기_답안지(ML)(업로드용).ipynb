````python
#ì •ë‹µ1(list comprehension)
alist = [x**2 for x in [2, 3, 4, 5, 6, 7] ]
print(alist)
````

````python
#ì •ë‹µ2
alist = []
for x in [2, 3, 4, 5, 6, 7]:
  alist.append(x**2)
print(alist)
````

### **1-3-b. ë¦¬ìŠ¤íŠ¸ "alist" ì¤‘ ë‹¤ìŒ ì¡°ê±´ì— ë§ëŠ” ìš”ì†Œë§Œ ì¶”ì¶œí•˜ì—¬, ìƒˆë¡œìš´ ë¦¬ìŠ¤íŠ¸ "blist"ë¥¼ ìƒì„±í•´ ì¶œë ¥í•´ ì£¼ì„¸ìš”. (3ì )**
- alist ì¤‘ ì•½ìˆ˜ê°€ (1)1, (2)ìì‹ ì˜ ì œê³±ê·¼, (3)ìê¸° ìì‹ ë§Œ ìˆëŠ” ìˆ«ìë§Œ ì°¾ì•„ ë¦¬ìŠ¤íŠ¸ "blist"ì— ë„£ì–´ ì£¼ì„¸ìš”.
- ì§ì ‘ ìˆ«ìë¥¼ ì°¾ì•„ ì…ë ¥í•˜ë©´ 0ì  ì²˜ë¦¬ë©ë‹ˆë‹¤.

````python
#ì •ë‹µ
blist = []
for n in alist:
    root = int(n**0.5)
    if root * root == n:  # ì œê³±ìˆ˜ì¸ì§€ í™•ì¸< ìƒëµ ê°€ëŠ¥
        # ì œê³±ê·¼ì´ ì†Œìˆ˜ì¸ì§€ í™•ì¸
        if root > 1 and all(root % i != 0 for i in range(2, int(root**0.5) + 1)):
            blist.append(n)
print( blist)
````

````python
def binary_iterative(num):
  result = "" # ë¬¸ìì—´ì— ë‚˜ë¨¸ì§€ë¥¼ ì €ì¥
  if num == 0:
    return "0"
  while num > 0:
    result = str(num%2) + result
    num = num//2 # ëª« update
  return result
````
