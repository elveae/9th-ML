### 15. What is the result of the following expression?

(실행 후, 결과에 대한 주석 작성해주세요. ex. 출력 결과에 대한 이유)

```python
0 * np.nan
np.nan == np.nan
np.inf > np.nan
np.nan - np.nan
np.nan in set([np.nan])
0.3 == 3 * 0.1
```

````python
print(0 * np.nan)   # nan: nan은 연산할 수 없다
print(np.nan == np.nan)   # False: nan은 어떤 값과도 같지 않다(자기자신 포함)
print(np.inf > np.nan)   # False: nan과의 비교는 항상 false
print(np.nan - np.nan)   # nan: nan은 연산할 수 없다
print(np.nan in set([np.nan]))   # True: set에 np.nan 객체가 저장돼 있으므로 동일한 객체로 인식
print(0.3 == 3 * 0.1)   # False: 부동소수점 연산 오차로 인해 정확히 같지 않음
````

### 21. Given a DataFrame of random numeric values:

````python
# 랜덤 시드 고정
np.random.seed(2025)

df = pd.DataFrame(np.random.random(size=(5, 3)))
df
````
